```python
import numpy as np
import matplotlib.pyplot as plt

# ----- Parameters -----
r = 2.5                     # logistic parameter
f = lambda y: r * y * (1 - y)

y0 = 0.8                    # initial condition (try 0.2, 0.4, 0.8, etc.)
T  = 30                     # number of iterations

# ----- Generate trajectory -----
ys = np.zeros(T+1)
ys[0] = y0
for t in range(T):
    ys[t+1] = f(ys[t])

# ----- Set up grid for plotting function and 45-degree line -----
y_grid = np.linspace(0, 1, 400)
f_vals = f(y_grid)

plt.figure(figsize=(6, 6))

# Plot the function y_{t+1} = f(y_t)
plt.plot(y_grid, f_vals, label=r"$y_{t+1} = f(y_t)$")

# Plot the 45-degree line y_{t+1} = y_t
plt.plot(y_grid, y_grid, linestyle="--", label=r"$y_{t+1} = y_t$")

# ----- Cobweb: vertical & horizontal segments -----
y_current = y0
for t in range(T):
    # vertical: (y_t, y_t) -> (y_t, f(y_t))
    plt.plot([y_current, y_current],
             [y_current, f(y_current)],
             linewidth=1)

    # horizontal: (y_t, f(y_t)) -> (f(y_t), f(y_t)) = (y_{t+1}, y_{t+1})
    plt.plot([y_current, f(y_current)],
             [f(y_current), f(y_current)],
             linewidth=1)

    y_current = f(y_current)

# Mark initial point
plt.scatter([y0], [0], color="black", zorder=5)
plt.text(y0, -0.03, r"$y_0$", ha="center")

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel(r"$y_t$")
plt.ylabel(r"$y_{t+1}$")
plt.title(r"Cobweb Diagram for $y_{t+1} = 2.5 y_t (1 - y_t)$")
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ----- Function -----
f = lambda y: 6*y - 6*y**2 - (2/3)

# ----- Initial condition and iterations -----
y0 = 0.7            # try values like 0.1, 0.2, 0.5, 0.7, 1.2, etc.
T  = 40             # number of cobweb iterations

# ----- Generate trajectory -----
ys = np.zeros(T+1)
ys[0] = y0
for t in range(T):
    ys[t+1] = f(ys[t])

# ----- Set up grid for plotting f(y) and the 45-degree line -----
# Choose a range that captures interesting dynamics.
# This function may go outside [0,1], so use a wider interval.
y_grid = np.linspace(-1.5, 1.5, 800)
f_vals = f(y_grid)

plt.figure(figsize=(7, 7))

# Plot the function y_{t+1} = f(y_t)
plt.plot(y_grid, f_vals, label=r"$y_{t+1} = f(y_t)$")

# Plot the 45-degree line y_{t+1} = y_t
plt.plot(y_grid, y_grid, linestyle="--", label=r"$y_{t+1} = y_t$")

# ----- Cobweb: vertical & horizontal segments -----
y_current = y0
for t in range(T):
    # vertical: (y_t, y_t) -> (y_t, f(y_t))
    plt.plot([y_current, y_current],
             [y_current, f(y_current)],
             color="k", linewidth=1)

    # horizontal: (y_t, f(y_t)) -> (f(y_t), f(y_t))
    plt.plot([y_current, f(y_current)],
             [f(y_current), f(y_current)],
             color="k", linewidth=1)

    y_current = f(y_current)

# Mark initial point
plt.scatter([y0], [0], color="black", zorder=5)
plt.text(y0, -0.1, r"$y_0$", ha="center")

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.xlabel(r"$y_t$")
plt.ylabel(r"$y_{t+1}$")
plt.title(r"Cobweb Diagram for $y_{t+1} = 6y_t - 6y_t^2 - \frac{2}{3}$")
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```
